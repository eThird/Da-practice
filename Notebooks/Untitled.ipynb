Using .groupby(), find the top 2 most sold coffees for each cash_type.

```python
top2=(new_cofee_Dataset.groupby(['cash_type','coffee_name']).size().reset_index(name='sales_count'))

top_2_results=(
    top2.sort_values(['cash_type','coffee_name'],ascending=[True, False]).groupby('cash_type').head(2)
)
top_2_results

```

Create a pivot table showing cash_type as rows, coffee_name as columns, and total money as values.

```python
pivoted_table = new_cofee_Dataset.pivot_table(
    index='cash_type',
    columns='coffee_name',
    values='money',
    aggfunc='sum'
)
pivoted_table
```
